```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\BCP\Downloads\results.csv')
data.head()
```

## Data Visualization

#### 1) Graph the maximum volatility trend per company (A single Line Chart: Each line refers to a company) 
#### Which company is the most volatile?- COSTCO

```python
plt.figure(figsize=(14, 10))
for company in data['name'].unique():
    c_data = data[data['name'] == company]
    c_data = c_data.sort_values('date')  # Sort data by date
    plt.plot(c_data['date'], c_data['max_volatility'], label=company, linewidth=2.5)

plt.title('Maximum Volatility per Company')
plt.xlabel('Date')
plt.ylabel('Highest Volatility')
plt.legend(title='Company', loc='upper right')
plt.show()
```

#### 2) Graph the daily average volatility per company (A Grouped Bar Chart: Each group refers to a company and the bars refer to the daily highest volatility)
#### Do the findings from this graph support your conclusion from the first graph? -Yes (COSTCO's average also seems to be the highest compared to other companies)

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=data, x='name', y='avg_volatility', hue='date')
plt.title('Daily Average Volatility per Company')
plt.xlabel('Company')
plt.ylabel('Average Volatility')
plt.legend(title='Date', loc='upper right')
plt.xticks(rotation=45)
plt.show()
```
